For each predictor, fit a simple linear regression model to predict the response. Describe your results. In which of the models is there a statistically significant association between the predictor and the response? Create some plots to back up your assertions. Are there any outliers that you would like to remove from your data for each of these regression tasks?


```python
def regression_plot(predictor):
    # Plot the data points
    plt.scatter(X[predictor], y, label='Data Points', color='blue', alpha=0.5)

    # Plot the regression line
    regression_line = model.predict(X)
    plt.plot(X[predictor], regression_line, label='Regression Line', color='red')

    # Calculate residuals
    residuals = y - regression_line
    
    # Standard deviation of residuals
    std_dev = residuals.std()

    # Identify potential outliers (e.g., residuals are more than 2 standard deviations from the mean)
    outliers = plant_df[np.abs(residuals) > 2 * std_dev]

    # Plot outliers
    plt.scatter(outliers[predictor], outliers['PE'], label='Outliers', color='yellow', s=50)

    # Additional plot formatting
    plt.title(f'Regression with Outliers for {predictor}')
    plt.xlabel(predictor)
    plt.ylabel('PE')
    plt.legend()
    plt.show()
```

```python
import statsmodels.api as sm

predictors = ['AT', 'V', 'AP', 'RH']
regression_results = {}

for predictor in predictors:
    X = plant_df[[predictor]]
    y = plant_df['PE']
    X = sm.add_constant(X)
    model = sm.OLS(y, X).fit()
    
    # Store results
    regression_results[predictor] = {
        'coefficient': model.params[predictor],
         'std_error': model.bse[predictor],
        'p-value': round(model.pvalues[predictor], 3),
        'r2': model.rsquared
    }
    regression_plot(predictor)
# Display the regression results
for predictor, stats in regression_results.items():
    print(f"{predictor}: Coefficient={stats['coefficient']:3f}, "
          f"Standard Error={stats['std_error']:.4f}, "
          f"P-value={stats['p-value']:.3f}, R2={stats['r2']:.3f}")

```
